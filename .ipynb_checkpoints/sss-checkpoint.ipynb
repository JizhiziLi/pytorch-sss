```python
def load_image_feat_array(image_path,feature_path):
    image = np.asarray(plt.imread(image_path))
    feat = np.asarray(scipy.io.loadmat(feature_path)['embedmap'])
    print(f'shape of image array is',image.shape)
    print(f'shape of feature array is',feat.shape)
    return image, feat

image, feature = load_image_feat_array("docia.png", "docia.mat")
image = plt.imread("docia.png")

# randomly show some images in the plot
f, axarr = plt.subplots(1,2)
axarr[0].imshow(image)
axarr[1].imshow(feature[:,:,random.sample(range(1, 128), 3)])

```

```python
# import cv2
# from cv2.ximgproc import guidedFilter

def pca_feature(image,feature):
    # Filter out super high numbers due to some instability in the network
    feature[feature>5] = 5
    feature[feature<-5] = -5
    #### Missing an image guided filter with the image as input
    ##
    ##########
    # change to double precision
    feature = np.float64(feature)
    # retrieve size of feature array
    shape = feature.shape
    [h, w, d] = feature.shape
    # resize to a two-dimensional array
    feature = np.reshape(feature, (h*w,d))
    # calculate average of each column
    featmean = np.average(feature,0)
    onearray = np.ones((h*w,1))
    featmeanarray = np.multiply(np.ones((h*w,1)),featmean)
    feature = np.subtract(feature,featmeanarray)
    feature_transpose = np.transpose(feature)
    cover = np.dot(feature_transpose, feature)
    # get largest eigenvectors of the array
    val,vecs = eigs(cover, k=3, which='LI')
    pcafeature = np.dot(feature, vecs)
    pcafeature = np.reshape(pcafeature,(h,w,3))
    pcafeature = np.float64(pcafeature)
    return pcafeature


def normalise_0_1(feature):
    max_value = np.amin(feature)
    min_value = np.amax(feature)
    subtract = max_value - min_value
    for i in range(0,3):
        feature[:,:,i] = feature[:,:,i] - np.amin(feature[:,:,i])
        feature[:,:,i] = feature[:,:,i] / np.amax(feature[:,:,i])
    return feature



pca_feature = pca_feature(image,feature)
normalise_feature = normalise_0_1(pca_feature)

f, axarr = plt.subplots(1,2)
axarr[0].imshow(image)
axarr[1].imshow(normalise_feature)

```

## Matting Affinity

```python
def localRGBnormalDistributions(image, windowRadius = None, epsilon = None):
    if windowRadius == None:
        windowRadius = 1
    if epsilon == None:
        epsilon = 1e-8
    [h, w, d] = image.shape
    N = h * w
    windowSize = 2 * windowRadius + 1
    meanImage = cv2.blur(image, (windowSize, windowSize))
    covarMat = np.zeros((3, 3, N),dtype='float')
    
    for r in range(0,3):
        for c in range(r,3):
            temp = cv2.blur(np.dot(image[:, :, r],image[:, :, c]), (windowSize,windowSize)) - np.dot(meanImage[:,:,r],meanImage[:,:,c])
            temp = np.reshape(temp, (409600,))
            covarMat[r, c, :] = temp[:]
            
    for i in range(0,3):
        covarMat[i,i,:] = covarMat[i,i,:] + epsilon
        
    for r in range(1,3):
        for c in range(0,r-1):
            covarMat[r,c,:] = covarMat[c,r,:]
            
    return meanImage, covarMat
    
# localRGBnormalDistributions(image,1,1e-8)

def im2col_sliding_strided(A, BSZ, stepsize=1):
    # Parameters
    m,n = A.shape
    s0, s1 = A.strides    
    nrows = m-BSZ[0]+1
    ncols = n-BSZ[1]+1
    shp = BSZ[0],BSZ[1],nrows,ncols
    strd = s0,s1,s0,s1

    out_view = np.lib.stride_tricks.as_strided(A, shape=shp, strides=strd)
    return out_view.reshape(BSZ[0]*BSZ[1],-1)[:,::stepsize]
```

```python
from scipy.sparse import csr_matrix

def calculate_matting_affinity(image):
    windowRadius = 1
    epsilon = 1e-7
    image = np.float64(image)
    windowSize = 2 * windowRadius + 1
    neighSize = windowSize**2
    [h, w, c] = image.shape
    N = h * w
    epsilon = epsilon / neighSize
    
    meanImage, covarMat = localRGBnormalDistributions(image, windowRadius, epsilon)
    
    # Determine pixels and their local neighbors
    indices = np.transpose(np.reshape(np.arange(1,h*w+1),(h,w)))
    neighInd = np.transpose(im2col_sliding_strided(indices,[windowSize, windowSize]))
    inMap = np.ones((h,w))
    inMap = inMap[windowRadius + 1 : inMap.shape[0]+1 - windowRadius, windowRadius + 1 : inMap.shape[1]+1 - windowRadius]
#     neighInd = neighInd[inMap, :]
    inInd = neighInd[:, int((neighSize +1)/2)]
    pixCnt = inInd.shape[0]
    print(f'pixCnt is ',pixCnt)
    print(f'inInd shape is ',inInd.shape)
    
    # Prepare in & out data
    image = np.reshape(image, (N, c))
    meanImage = np.reshape(meanImage, (N, c))
    flowRows = np.zeros((neighSize, neighSize, pixCnt),dtype='float')
    flowCols = np.zeros((neighSize, neighSize, pixCnt),dtype='float')
    flows = np.zeros((neighSize, neighSize, pixCnt),dtype='float')
    
    for i in range(0,inInd.shape[0]-1):
        if(i%100000==0):
            print(i)
        neighs = neighInd[i,:]
        shiftedWinColors = image[neighs,:] - np.matlib.repmat(meanImage[inInd[i],:],neighs.shape[0],1)
        arraydivide = np.dot(np.linalg.inv(covarMat[:,:,inInd[i]]),np.transpose(shiftedWinColors))
        flows[:,:,i] = np.dot(shiftedWinColors, arraydivide)
        neighs = np.matlib.repmat(neighs,neighs.shape[0],1)
        flowRows[:,:,i] = neighs
        flowCols[:,:,i] = np.transpose(neighs)
        
    flows = (flows+1)/neighSize
    W = csr_matrix((flows[:].flatten(), (flowRows[:].flatten(), flowCols[:].flatten())), shape=(N, N))
    print(W.shape)
    
        



    
calculate_matting_affinity(image)
```

## Generate superpixal

```python
from skimage.segmentation import slic
from skimage.segmentation import mark_boundaries
from skimage.util import img_as_float
from skimage import io
import matplotlib.pyplot as plt
import argparse
```

```python
image = img_as_float(image)
# loop over the number of segments
for numSegments in (100, 200, 300):
    # apply SLIC and extract (approximately) the supplied number
    # of segments
    segments = slic(image, n_segments = numSegments, sigma = 5)
     # show the output of SLIC
    fig = plt.figure("Superpixels -- %d segments" % (numSegments))
    ax = fig.add_subplot(1, 1, 1)
    ax.imshow(mark_boundaries(image, segments))
    plt.axis("off")
    
plt.show()
```
